```python
details=['TV','Radio','Newspaper','sales']
```

```python
for column in details:
    print(df[column].value_counts())
    plt.figure(figsize=(10,7))
    sns.countplot(x=df[column])
    plt.xticks(rotation=45)
    plt.show()
    
```
